```python
from collections import deque

class Node:

    def __init__(self, x):

        self.key = x
        self.left = None
        self.right = None

minValue, maxValue = 0, 0

def findHeight(node):

    if (node == None):
        return 0
    leftHeight = findHeight(node.left)
    rightHeight = findHeight(node.right)


    if leftHeight > rightHeight:
        return leftHeight + 1

    return rightHeight + 1


def isPalindrome(str):
    l = 0;
    h = len(str) - 1


    while (h > l):
        if (str[l] != str[h]):
            return False
        l += 1
        h -= 1

    return True

def isPathPal(path, index):

    i = 0
    s = ""

    while (i <= index):
        s += str(path[i])
        i += 1
    return isPalindrome(s)

def printPalPath(path,index):


    for i in range(index):
        print(path[i], end = ", ")
    print()

def printPath(node, path, index):


	if (node == None):
		return


	path[index] = node.key;

	printPath(node.left,
			path, index + 1)

	printPath(node.right, 
			path, index + 1)


	if (node.left == None and
		node.right == None):

		if (isPathPal(path, index)):
			printPalPath(path, 
						index + 1)

def PalindromicPath(node):

	# Calculate the height
	# of the tree
	height = findHeight(node)
	path = [0] * height

	printPath(node, path, 0)

def createAndPrintPalPath():

	# /*	 2
	#	 / \
	#	 6	 8
	#		 / \
	#	 8 5
	#	 / \ / \
	#	 1 2 3 8
	#		 /
	#		 2
	# */

	# Creation of tree
	root = Node(2)
	root.left = Node(6)
	root.right = Node(8)

	root.right.left = Node(8)
	root.right.right = Node(5)

	root.right.left.left = Node(1)
	root.right.left.right = Node(2)
	root.right.right.left = Node(3)
	root.right.right.right = Node(8)
	root.right.right.right.left = Node(2)

	PalindromicPath(root)

if __name__ == '__main__':

	createAndPrintPalPath()

```
